```python
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
def load_data(path, curvatures):
    data = {}
    for c in curvatures:
        data[c] = np.load(path / f'gi_data_curv_{c}.npy', allow_pickle=True)
    return data
```

```python
curvatures = [0, 5, 10, 15, 20]
fiber_data = load_data(Path('mmf'), curvatures)
```

```python
for k in fiber_data:
    fiber_data[k][fiber_data[k] > 1] = 0
    plt.imshow(fiber_data[k])
    plt.show()
```

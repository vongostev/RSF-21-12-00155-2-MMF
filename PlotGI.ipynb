```python
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
def load_data(path, curvatures):
    data = {}
    for c in curvatures:
        data[c] = np.load(path / f'gi_data_curv_{c}.npy', allow_pickle=True)
    return data
```

```python
curvatures = [0, 5, 10, 15, 20]
fiber_data = load_data(Path('mmf'), curvatures)
```

```python
fig, axes = plt.subplots(1, len(curvatures), figsize=(16, 3), dpi=300)
for i, k in enumerate(curvatures):
    fiber_data[k][fiber_data[k] > 1] = 0
    axes[i].imshow(fiber_data[k])
    axes[i].set_title(f'Curvature {k} cm')
```

```python
fdata = np.load(f'mmf/gi_data_curv_20_straight.npy', allow_pickle=True)
plt.imshow(fdata)
plt.title(f'Curvature 20 cm')
plt.colorbar()
```

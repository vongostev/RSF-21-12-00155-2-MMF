```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_modes(modes, indexes, bounds):
    npoints = np.ceil(np.sqrt(len(modes[0]))).astype(int)
    fig, axes = plt.subplots(2, 5, dpi=100, figsize=(10, 4))
    for mode, ax in zip(modes[indexes], axes.flatten()):
        ax.imshow(np.real(mode).reshape((npoints, -1)),#[npoints // 4:- npoints // 4, npoints // 4:- npoints // 4],
                extent=bounds * 2)
    plt.tight_layout()
    # plt.savefig('pcf/pcf_modes.png', dpi=100)
    # plt.savefig('simc/smf_simc_modes.png', dpi=100)
    plt.show()
```

```python
bounds = [-31.25 * 3.5 / 2, 31.25 * 3.5 / 2]
```

```python
loaded_data = np.load('data/mmf_GRIN_62.5_properties_0.npz', mmap_mode='r')
# loaded_data = np.load('mmf_PCF_3.6_properties.npz')
modes = loaded_data['modes_list']
plot_modes(modes, slice(-11, -1), bounds)
```

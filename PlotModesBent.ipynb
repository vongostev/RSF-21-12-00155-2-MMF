```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_modes(modes, indexes, bounds):
    npoints = np.ceil(np.sqrt(len(modes[0]))).astype(int)
    indexes_map = np.arange(len(modes))[indexes]
    fig, axes = plt.subplots(2, 5, dpi=100, figsize=(10, 4))
    for modnum, mode, ax in zip(indexes_map, modes[indexes], axes.flatten()):
        ax.imshow(np.abs(mode).reshape((npoints, -1)) ** 2, extent=bounds * 2)
        ax.set_title(f'Mode No. {modnum}')
    plt.tight_layout()
    # plt.savefig('pcf/pcf_modes.png', dpi=100)
    # plt.savefig('simc/smf_simc_modes.png', dpi=100)
    plt.show()
```

```python
bounds = [-31.25 * 3.5 / 2, 31.25 * 3.5 / 2]
modes_indices = np.arange(-11, 0, 1)
```

Распределения интенсивности 10 старших мод для прямого волокна

```python
loaded_data = np.load('data/mmf_GRIN_62.5_properties_0.npz', mmap_mode='r', allow_pickle=True)
modes = loaded_data['modes_list']
plot_modes(modes, modes_indices, bounds)
```

Распределения интенсивности 10 старших мод для волокна с радиусом кривизны $R=20$ см
